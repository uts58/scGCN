```python
import matplotlib.pyplot as plt
import pandas

base_path = 'C:\\Users\\uts58\\OneDrive - North Dakota University System\\masters_research\\Plot\\liu_GSE223917'

df_scanpy = pandas.read_csv(f'{base_path}\\brain new\\GSE223917_HiRES_brain_scanpy.csv')
df_X = pandas.read_excel(f'{base_path}\\brain new\\HiC + RNA Signle Chr (7 to 14)\\brain_hic_rna.xlsx')
df = pandas.read_excel(f'{base_path}\\brain new\\HiC Single Chr (7 to 14)\\brain_hic.xlsx')

# Create a single figure for three separate subplots
fig = plt.figure(figsize=(15, 5))

# Subplot 1
ax1 = fig.add_subplot(131)
ax1.boxplot(df_X['adjusted_rand'])
ax1.set_title('HiC + RNA')
ax1.set_ylabel('Adjusted Rand Index')

# Subplot 2
ax2 = fig.add_subplot(132)
ax2.boxplot(df['adjusted_rand'])
ax2.set_title('HiC')
ax2.set_ylabel('Adjusted Rand Index')

# Subplot 3
ax3 = fig.add_subplot(133)
ax3.boxplot(df_scanpy['adjusted_rand'])
ax3.set_title('RNA')
ax3.set_ylabel('Adjusted Rand Index')

# Show the plot
plt.tight_layout()
plt.savefig(f'{base_path}\\brain new\\GSE223917_brain_ari.png')
plt.show()

fig = plt.figure(figsize=(15, 5))

# Subplot 1
ax1 = fig.add_subplot(131)
ax1.boxplot(df_X['silhouette_avg'])
ax1.set_title('HiC + RNA')
ax1.set_ylabel('Silhouette Average')

# Subplot 2
ax2 = fig.add_subplot(132)
ax2.boxplot(df['silhouette_avg'])
ax2.set_title('HiC')
ax2.set_ylabel('Silhouette Average')

# Subplot 3
ax3 = fig.add_subplot(133)
ax3.boxplot(df_scanpy['silhouette_avg'])
ax3.set_title('RNA')
ax3.set_ylabel('Silhouette Average')

# Save the plot
plt.tight_layout()
plt.savefig(f'{base_path}\\brain new\\GSE223917_brain_silhouette.png')

# Display the saved plot
plt.show()
```
